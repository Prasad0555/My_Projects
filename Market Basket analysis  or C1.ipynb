```python
import pandas as pd
```

```python
Transactions = [['Milk','Butter','Egg','Bread','Cerelac'],
               ['Milk','Egg','Bread'],
               ['Milk','Bread'],
               ['Egg','Apple','Bread'],
               ['Cerelac','Diaper','Babysoap'],
               ['Cerelac','Diaper','Egg','Butter','Bread'],
               ['Apple','Cerelac','Milk']]

# here we are created 7 transaction. As of now this is a list of transaction we convert this into an array then we will
# convery into dataframe
```

```python
!pip install mlxtend
from mlxtend.preprocessing import TransactionEncoder
```

```python
TE = TransactionEncoder()

TE_array = TE.fit(Transactions).transform(Transactions)

# Create a boolean table for the items included/not included in the transaction
transac = pd.DataFrame(TE_array, columns=TE.columns_)

# here we transform this transactions into a DataFrame. First we convert then into an array then using the array we convert 
# them into a DataFrame. Because we want to convert the transaction list into a list of boolean values.
```

```python
,
# so here wherever these items is present it is true and wherever these items is not present it is false.
# here we have created a boolean table for the above transaction.
```

```python
from mlxtend.frequent_patterns import apriori

freq_itemsets = apriori(transac, min_support=0.01, use_colnames=True)
```

```python
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
print(freq_itemsets)

# here you are seeing is first one-item-itemset then two-item-itemset then three-item-itemset, four-item-itemset and 
# five-item-itemset.
# Each itemset have the support values which simply tells you the proportion of times this particular item has occured in
# the transaction list.
```

```python
from mlxtend.frequent_patterns import fpgrowth

frequent_itemsets_fp = fpgrowth(transac, min_support=0.01, use_colnames=True)

freq_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

rules_ap = association_rules(freq_itemsets, metric="confidence", min_threshold=.8)

# having found the frequent item itemset now we are minning the pattern, now when we are minning the pattern we are using 
# the metric to be the confidence.
# In output: antecedents have the (Babysoap) and consequents have the (Cerelac) it simply says all those people who brought 
#            Babysoap they also brought Cerelac.
```

```python
 #Explanation of the term Lift

Lift=(support of A,B)/((Support of A)*(Support of B))

Lift value of 1 indicated that the items A and B are independent

Lift value >1 indicates that the purchase of product A will lead to a purchase of product B too
```
