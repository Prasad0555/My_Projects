```python
# Import library for VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):
    
    # calculation VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    
    return(vif)
```

```python
calc_vif(df1) # there is no multicollinearity in the variables
```

# Scale our numeric variables using min-max normalization

```python
# apply the min-max scaling to our numeric variables
from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler()

# Scaling down the numeric variables
df = pd.DataFrame(min_max.fit_transform(df1.iloc[:,0:7]),
columns = df1.iloc[:,0:7].columns.tolist())
```

```python
df
```

```python
df.describe()
```

```python
df.shape
```

```python
# create feature and response variable set
x = df.drop(['house price of unit area'], axis=1)   # Under X we are taking all the input variables
y = df['house price of unit area']    # Under Y we are just taking the target variable 
```

# Create training and testing datasets using the train_test_split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3)  # we are splitting the ration is 70:30
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
#Fitting the MLR model to the training set:  
from sklearn.linear_model import LinearRegression  
regressor= LinearRegression()  
regressor.fit(x_train, y_train) 
```

```python
#Predicting the Test set result;  
y_pred= regressor.predict(x_test) 
```

```python
y_pred  # this is my predicted values
```

```python
print('Train Score: ', regressor.score(x_train, y_train))  
print('Test Score: ', regressor.score(x_test, y_test))  
```
