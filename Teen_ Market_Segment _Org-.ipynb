Also the age column has 5086 missing values. One way to deal with these missing values would be to fill the missing values
with the average age of each graduation year.

```python
data.groupby('gradyear')['age'].mean()
```

From the above summary we can observe that the mean age differs by roughly one year per change in graduation year. This is 
not at all surprising, but a helpful finding for confirming our data is reasonable.

```python
# We now fill the missing values for each graduation year with the mean that we got as above
data['age']= data.groupby('gradyear').transform(lambda x : x.fillna (x.mean()))
```

# Treating Outliers

The original age range contains value from 3 - 106, which is unrealistic because students at age of 3 or 106 would not attend 
high school. A reasonable age range for people attending high school will be the age range between 13 to 2. The rest should be
treated as outliers keeping the age of student going to high school in mind. Lets detect teh outliers using a box plot below

```python
sns.boxplot(data['age'])
```

```python
q1= data['age'].quantile(0.25)
q3= data['age'].quantile(0.75)
iqr= q3-q1
```

```python
print(iqr)
```

```python
df= data[(data['age'] > (q1 - 1.5*iqr)) & (data['age'] < (q3 + 1.5*iqr))]
```

```python
df['age'].describe()
```

```python
df.shape 
```

```python
sns.boxplot(df['age'])
```

# Convert object variable to Numeric

```python
def gender_to_numeric(x):
    if x=='M':
        return 1
    if x=='F':
        return 2
    if x=='not disclosed':
        return 3
    
```

```python
scaled_feature['gender'] = scaled_feature['gender'].apply(gender_to_numeric)
scaled_feature['gender'].head()
```

# Elbow Method

The location of a bend(knee) in the plot is generally considered as an indicator of the appropriate number of clusters.
Our Elbow point is around cluster size of 5. We will use k=5 to further interpret our clustering result.

We are trying to group in teen agers according to the similarity which is getting exhibited by the words which they are using
on their social networking platform.

# Fit K-Means clustering for k=5

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(scaled_feature)
```

As a result of clustering, we have the clustering label. Lets put these labels back into the original numeric data frame.

```python
len(kmeans.labels_)
```

```python
data['cluster']=kmeans.labels_
```

```python
data.head()
```
